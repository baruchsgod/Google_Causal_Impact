```python
stocks = ['META', 'WMT', 'DIS', 'BMW.DE', 'NVS', 'GS', 'GE', 'KHC', 'MCD', 'CARL-B.CO']
```

```python
data = yf.download(stocks, start = start, end = end_stock, interval = "1wk" )
```

```python
data = data['Adj Close']
```

```python
data = data.dropna()
```

```python
data_cor = data[data.index < treatment_start]
```

```python
data_cor.corr()
```

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
#Create dataset with the most correlated stocks
final_series = data[['META','CARL-B.CO', 'MCD', 'GS','WMT', 'NVS']]
```

```python
#Create pre and post period
pre_period = [start, training_end]
post_period = [treatment_start, end]
```

```python
from causalimpact import CausalImpact
```

```python
impact = CausalImpact(data = final_series,
                      pre_period = pre_period,
                      post_period = post_period,
                      nseasons = [{'period':52}],
                      priod_level_sd = None)
```

```python
impact.plot()
```

```python
print(impact.summary())
```

```python
print(impact.summary('report'))
```
